```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'records.csv'
df = pd.read_csv(file_path)
```

```python
# Hapus kolom yang semua isinya berisi NaN
df_cleaned = df.dropna(axis=1, how='all')
```

```python
df_cleaned.head()
```

```python
# Filter data untuk emotion scenario
emotion_df = df_cleaned[df_cleaned['record_type.name'].isin(['I-emotion'])]
emotion_df.head()
```

```python
# Mendapatkan nilai-nilai unik dari kolom 'record_type.name'
unique_activity_types = emotion_df['activity_type.name'].unique()
unique_activity_types
```

```python
# Filter kolom yang diperlukan
filtered_df = emotion_df[['value', 'activity_type.name']]
```

```python
# Filter dan hapus baris dengan nilai NaN pada kolom 'value'
filtered_df.dropna(subset=['value'], inplace=True)
```

```python
# Plotting bar chart
plt.figure(figsize=(10, 6))
plt.bar(filtered_df['activity_type.name'], filtered_df['value'])
plt.xlabel('Activity Type')
plt.ylabel('Value')
plt.title('Bar Chart of Value vs Activity Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
